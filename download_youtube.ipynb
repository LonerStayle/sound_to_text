```python
import yt_dlp

ydl_opts = {'format': 'mp4'}
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download(['https://www.youtube.com/watch?v=rOreQ-3De6I'])
```

```python
import subprocess
from pathlib import Path

VIDEO_PATH = 'asdasd.mp4'
OUTPUT_DIR = Path('./output')
OUTPUT_DIR.mkdir(exist_ok=True)

SEGMENT = 6
DURATION = 300
padding = len(str(DURATION // SEGMENT))  # 파일 번호 padding 길이 자동 계산, 여기선 2자리

for idx, start in enumerate(range(0, DURATION, SEGMENT), start=1):
    filename = f"output{idx:0{padding}d}"
    audio_out = OUTPUT_DIR / f"{filename}.wav"
    img_out = OUTPUT_DIR / f"{filename}.jpg"

    # 오디오 WAV 추출
    subprocess.run([
        'ffmpeg', '-y',
        '-i', VIDEO_PATH,
        '-ss', str(start),
        '-t', str(SEGMENT),
        '-vn',
        '-acodec', 'pcm_s16le',
        '-ar', '44100',
        '-ac', '2',
        str(audio_out)
    ], check=True)

    # 지정 시점 프레임 캡처
    subprocess.run([
        'ffmpeg', '-y',
        '-i', VIDEO_PATH,
        '-ss', str(start),
        '-frames:v', '1',
        str(img_out)
    ], check=True)

print("모든 세그먼트의 WAV 및 이미지 저장 완료!")
```
